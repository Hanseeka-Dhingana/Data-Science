#### Scrape Data from the Daraz.com website 
 - First get the data from the daraz.com   

```python
import requests
from bs4 import BeautifulSoup as sp


getData = requests.get("https://www.daraz.pk/#hp-just-for-you")

print(getData.status_code)

content = getData.content
print(content)

print("After Prettify content..................... \n ")
soup = sp(content, "html.parser")

text = soup.prettify()
print(text)
```

### Get title

```python
title = soup.title
title 
```

### Get the product list

```python
product_containers = soup.find_all('div', {'data-qa-locator': 'product-item'}) # This locator may need adjustment

products = []
for container in product_containers:
    # Extract desired information (adjust selectors based on actual HTML)
    try:
        name = container.find('div', class_='product-title-class').text.strip()
        price = container.find('span', class_='product-price-class').text.strip()
        # Add other fields like rating, link, etc.
        products.append({'Name': name, 'Price': price})
        
    except AttributeError:
        continue # Handle cases where an element might be missing
```
