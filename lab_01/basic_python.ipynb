Find the mean, mode, median of list using statistics library

```python
import statistics as st   

list = [1,2,3,4,5,3,3,2,1]

mode = st.mode(list)

median = st.median(list)

mean = st.mean(list)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
```

Convert list in to array, then find means, mode, median of array using numpy library.

```python
import numpy as np

list = [1,2,3,4,5,3,3,2,1]

array = np.array(list)

print("List to Array:", array)

arrayMean = np.mean(array)

arrayMedian = np.median(array)

arrayMode = st.mode(array)

print("Array Mean:", arrayMean)
print("Array Median:", arrayMedian)
print("Array Mode:", arrayMode)
```

#### Control structure in python 

Three types of control structure in python.
1) Sequential 
2) Selectional
3) Iteration

Example: (if,else, elif), match case, loop

by default the flow of program is sequential.

1) Generate the table of number using If else program  and loop


```python
num = int (input("Enter a number to generate table: "))

if(num>0):
    print("Table of ",num)
    for i in range(1, 11):
        print(num,"x",i,"=",num*i)
        
else:
    print("Please enter a positive integer")        
        
```

2) Calculator Program using Match Case.

```python
num1 = int(input("Enter first number"))

num2 = int(input("Enter second number"))

operation =input("Enter operation")

match(operation):
    case '+': print(f"Addition of {num1} and {num2} is {num1+num2}")
    case '-': print(f"Subtraction of {num1} and {num2} is {num1-num2}")
    case '*': print(f"Multiplication of {num1} and {num2} is {num1*num2}")
    case '/': print(f"Division of {num1} and {num2} is {num1/num2}")
    case _: print("Invalid operation")
    
```

3) Multiple Table of number using Loop

```python
num = int (input("Enter a number to generate table: "))

if(num>0):
    for i in range (num , 11):
      print("Table of ",num)
      for i in range(1, 11):
          print(num,"x",i,"=",num*i)      
      print("**********************")
      num = num+1
        
else:
    print("Please enter a positive integer")        
        
```

3) Fabonocies Program Series Using function 

```python
num = int(input("Enter first number"))


def fibnocci(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b
```

2) Factorial Program Using function

```python
num = int (input("Enter a number "))

def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n-1)
    
    
if num < 0:
    print("Factorial is not defined for negative numbers")   
     
else:
    print(f"Factorial of {num} is {factorial(num)}")
    
```
