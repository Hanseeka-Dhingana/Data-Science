## Simple Imputation

### 1) Mean Imputation

```python
# Simple Impute
from sklearn.impute import SimpleImputer
import numpy as np

# Numeric example
# Create Dataset
data = [ [2, 4, np.nan],
         [8, np.nan, 12],
         [np.nan, 16 ,18]]

# Imputation object
# use means imputation
imputer = SimpleImputer(strategy="mean")
data_imputed = imputer.fit_transform(data)

print(data_imputed)
```

### 2) Median Imputation

```python
# Imputation object
# use means imputation
imputer = SimpleImputer(strategy="median")
data_imputed = imputer.fit_transform(data)

print(data_imputed)
```

### 3) Most Frequent Imputation (Mode Imputation)

```python
# Imputation object
# use means imputation
imputer = SimpleImputer(strategy="most_frequent")
data_imputed = imputer.fit_transform(data)

print(data_imputed)
```

## KNN Imputation 

```python
# KNNImpute
from sklearn.impute import KNNImputer
import numpy as np


# create dataset
data = [ [2, 4, np.nan],
         [8, np.nan, 12],
         [np.nan, 16 ,18]]

# Imputation object assign NNeighrest Neighbour 
imputer = KNNImputer(n_neighbors=2)
data_imputed = imputer.fit_transform(data)

print(data_imputed)
```
