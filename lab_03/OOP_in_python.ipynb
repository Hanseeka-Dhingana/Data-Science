### Object Oriented Programming in the python

#### Create the class and two methods in the class
- Call this method through object of the class.
- In Python, methods inside a class are meant to work with an object (instance).
- self represents the current object that is calling the method.

```python
# class definition
class Student:
    
    num1 = 3
    num2 = 4

    # method definitions
    def first(self):
        print("This is the first method.")
        
    def second(self):
        print("This is the second method.")
        

obj = Student()
obj.first()
obj.second()

        
```

#### Create the constructor in the class

```python
# class definition
class Student:
    
    num1 = 0
    num2 = 0
    
    def __init__(self):
            print("This is the constructor method.")
            self.num1 = 3
            self.num2 = 4
            
    # method definitions
    def first(self):
        print("This is the first method. num 1 = ", self.num1)
        
    def second(self):
        print("This is the second method.num 2  = ", self.num2)
        

obj = Student()
obj.first()
obj.second()

        
```

#### Give the value to variable through class object

```python
# class definition
class Student:
    
    num1 = 0
    num2 = 0
    
    def __init__(self, num1, num2):
            print("This is the constructor method.")
            self.num1 = num1
            self.num2 = num2
            
    # method definitions
    def first(self):
        print("This is the first method. num 1 = ", self.num1)
        
    def second(self):
        print("This is the second method.num 2  = ", self.num2)
        

obj = Student(3,10)
obj.first()
obj.second()

        
```

#### Inheritance the class student2 from student

```python
class Teacher():
    
    def third(self):
        print("This is the method from Teacher class")
        

class Student(Teacher):
    def fourth(self): 
        print("This is the method from Student class")
        
        

teacher = Teacher()
teacher.third()

print("\n This is the object of Student class")
student = Student()
student.third()
student.fourth()
```
